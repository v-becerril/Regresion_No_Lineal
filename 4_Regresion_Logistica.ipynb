```python
#Cargamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Cargar archivo csv desde seaborn
df=sns.load_dataset(name='titanic')
df.head(20)
```

```python
#Convertir archivo filtrado a CSV
#df.to_csv("Titanic_Completo.csv")
```

```python
#Rellenamos nulos
df =df.fillna(method="bfill")
df =df.fillna(method="ffill")
```

```python
#Declaramos las variables dependientes e independientes para la regresión Logística
Vars_Indep= df[['pclass', 'age', 'fare']] #deben de ser númericas
Var_Dep= df['alive']
```

```python
#Redefinimos las variables 
X= Vars_Indep
y= Var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =None)
```

```python
#Se escalan todos los datos
#Escalar los datos es colocar un mismo rango y solo se escala los valores independientes
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las variables “X” tanto de entrenamiento como de prueba, utilizaremos fit_transform
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamos una predicción
y_pred = algoritmo.predict(X_test) 
y_pred
```

```python
#Verifico la matriz de Confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred, average="binary", pos_label="yes")
print('Precisión del modelo:')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred, average="binary", pos_label="yes")
print('Sensibilidad del modelo:')
print(sensibilidad)
```

```python
# el puntaje F1 que es una combinación entre la precisión y la sensibilidad, para esto importamos f1_score.
#Calculo el Puntaje F1 del modelo
from sklearn.metrics import f1_score

puntajef1 = f1_score(y_test, y_pred, average="binary", pos_label="yes")
print('Puntaje F1 del modelo:')
print(puntajef1)
```
